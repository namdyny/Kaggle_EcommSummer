```python
# import packages
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
# read csvs to DataFrames
uni_cat_df = pd.read_csv(
    'data/unique-categories.csv'
)
uni_cat_by_count_df = pd.read_csv(
    'data/unique-categories.sorted-by-count.csv'
)
product_master_df = pd.read_csv(
    'data/summer-products-with-rating-and-performance_2020-08.csv'
)
```

```python
# print all headers
uni_cat_index = uni_cat_df.columns
uni_cat_by_count_index = uni_cat_by_count_df.columns
product_master_index = product_master_df.columns
print(uni_cat_index)
print(uni_cat_by_count_index)
print(product_master_index)
```

```python
# histograms on prices
retail_price_hist = product_master_df['retail_price'].hist(bins=15)
price_hist = product_master_df['price'].hist(bins=15)
```

```python
# product color distribution
product_color_pie = product_master_df['product_color'].value_counts().plot.pie(y='mass', figsize=(5, 5))
```

```python
# bar on rating counts
count_five_star = product_master_df['rating_five_count'].sum()
count_four_star = product_master_df['rating_four_count'].sum()
count_three_star = product_master_df['rating_three_count'].sum()
count_two_star = product_master_df['rating_two_count'].sum()
count_one_star = product_master_df['rating_one_count'].sum()
rating_count_df = pd.DataFrame({
    'rating': ['one_star', 'two_star', 'three_star', 'four_star', 'five_star'],
    'count': [count_one_star, count_two_star, count_three_star, count_four_star, count_five_star],
})
rating_count_bar = rating_count_df.plot.bar(x='rating', y='count', rot=0)
```

```python
# histogram on merchant rating
merchant_rating_hist = product_master_df['merchant_rating'].hist(bins=15)
```

```python
# scatter plot on unit sold vs rating
scatter_unitsold_rating = product_master_df.plot.scatter(x='rating',
                      y='units_sold',
)
```
